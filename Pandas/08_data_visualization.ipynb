## 1. Import Libraries

```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
```

## 2. Load Dataset


```python
# import titanic data from seaborn library
df = sns.load_dataset('titanic')
```

## 3. Visualize the Data

---
- scatter plot

```python
df.plot(kind= 'scatter', x= 'age', y= 'fare')
```

> Note: fig # 01 This is shows the chart of age and fare

---
- scatter plot 

```python
df.plot( kind = 'scatter', x= 'age', y = 'fare', color = 'red')
```

> Note: fig 2 this plot shows the chart of age and fare with red color.

---
- Scatter plot

```python
# change colors using hex codes
df.plot( kind = 'scatter', x= 'age', y = 'fare', color = '#F67B0E')
```

> Note: fig 3 this plot shows the chart of age and fare with hexa color.

---
- Bar plot

```python
df.groupby('sex')['age'].mean().plot(kind = 'bar')
```

> Note: fig 4 this chart shows females are more than male.

- Bar chart

```python
df.groupby('sex')['fare'].mean().plot(kind = 'bar')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

> Note: fig 5 shows that female average fare is greater than male average fare

```python
import plotly.graph_objects as go

# Read data from a csv
z_data = pd.read_csv( 'https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title = 'Mt Bruno Elevation', autosize = False, width = 500, height = 500, margin = dict( l=65, r=50, b=65, t=90))

fig.show()
```
